```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
log_path = Path("/home/neil/repos/thesis/logs/")

# Single training
train_single_18 = log_path / "train_single_resnet18"
train_single_34 = log_path / "train_single_resnet34"

# Multiple training 50-50
train_mul_18 = log_path / "queues"
train_mul_34 = log_path / "queues"
```

```python
# RN18 by itself
df_18 = pd.read_csv(train_single_18 / "resnet18_pretrained_pid_65006.csv")

# RN34 by itself
df_34 = pd.read_csv(train_single_34 / "resnet34_pretrained_pid_66563.csv")

# 50-50 (no MPS, single)
df_sin_18 = pd.read_csv(train_single_18 / "resnet18_pretrained.csv")
df_sin_34 = pd.read_csv(train_single_34 / "resnet34_pretrained.csv")

# 40-60 (MPS, single)
df_sin_18_40p = pd.read_csv(train_single_18 / "resnet18_pretrained_pid_74282.csv")
df_sin_34_60p = pd.read_csv(train_single_34 / "resnet34_pretrained_pid_74283.csv")

# 50-50 (no MPS)
df_mul_18_50p = pd.read_csv(train_mul_18 / "resnet18_pretrained_bs50_4tw_4vw_1pf_pid_61135.csv")
df_mul_34_50p = pd.read_csv(train_mul_34 / "resnet34_pretrained_bs50_4tw_4vw_1pf_pid_61137.csv")

# 50-50 (no MPS), queue size 20
df_mul_18_50p_q20 = pd.read_csv(train_mul_18 / "resnet18_pretrained_bs50_4tw_4vw_1pf_pid_91744.csv")
df_mul_34_50p_q20 = pd.read_csv(train_mul_34 / "resnet34_pretrained_bs50_4tw_4vw_1pf_pid_91748.csv")

# 40-60 (MPS)
df_mul_18_40p = pd.read_csv(train_mul_18 / "resnet18_pretrained_bs50_4tw_4vw_1pf_pid_65865.csv")
df_mul_34_60p = pd.read_csv(train_mul_34 / "resnet34_pretrained_bs50_4tw_4vw_1pf_pid_65866.csv")

# 40-60 (MPS), queue size 20
df_mul_18_40p_q20 = pd.read_csv(train_mul_18 / "resnet18_pretrained_bs50_4tw_4vw_1pf_pid_94581.csv")
df_mul_34_60p_q20 = pd.read_csv(train_mul_34 / "resnet34_pretrained_bs50_4tw_4vw_1pf_pid_94582.csv")

# 35-65 (MPS)
df_mul_18_35p = pd.read_csv(train_mul_18 / "resnet18_pretrained_bs50_4tw_4vw_1pf_pid_69989.csv")
df_mul_34_65p = pd.read_csv(train_mul_34 / "resnet34_pretrained_bs50_4tw_4vw_1pf_pid_69990.csv")
```

```python
a = pd.DataFrame((df_18["total_time"] + df_34["total_time"]).cumsum()).reset_index()
a.columns = ["epochs", "acc_time"]
a["epochs"] += 1
#a["epochs"] = pd.Series([i for i in range(1,11)])
```

```python
df_18["acc_time"] = df_18["total_time"].cumsum()
df_34["acc_time"] = df_34["total_time"].cumsum()

df_sin_18["acc_time"] = df_sin_18["total_time"].cumsum()
df_sin_34["acc_time"] = df_sin_34["total_time"].cumsum()

df_sin_18_40p["acc_time"] = df_sin_18_40p["total_time"].cumsum()
df_sin_34_60p["acc_time"] = df_sin_34_60p["total_time"].cumsum()

df_mul_18_50p["acc_time"] = df_mul_18_50p["total_time"].cumsum()
df_mul_34_50p["acc_time"] = df_mul_34_50p["total_time"].cumsum()

df_mul_18_50p_q20["acc_time"] = df_mul_18_50p_q20["total_time"].cumsum()
df_mul_34_50p_q20["acc_time"] = df_mul_34_50p_q20["total_time"].cumsum()

df_mul_18_40p["acc_time"] = df_mul_18_40p["total_time"].cumsum()
df_mul_34_60p["acc_time"] = df_mul_34_60p["total_time"].cumsum()

df_mul_18_40p_q20["acc_time"] = df_mul_18_40p_q20["total_time"].cumsum()
df_mul_34_60p_q20["acc_time"] = df_mul_34_60p_q20["total_time"].cumsum()

#df_mul_18_35p["acc_time"] = df_mul_18_35p["total_time"].cumsum()
#df_mul_34_65p["acc_time"] = df_mul_34_65p["total_time"].cumsum()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
df_18.plot(x="acc_time", y="epoch", ax=ax, label="RN18", c="red", linewidth=2, alpha=1)
df_34.plot(x="acc_time", y="epoch", ax=ax, label="RN34", c="blue", linewidth=2, alpha=1)

a.plot(x="acc_time", y="epochs", ax=ax, label="RN18, RN34, combined")

df_sin_18.plot(x="acc_time", y="epoch", ax=ax, label="RN18", c="red", linewidth=5, alpha=0.3)
df_sin_34.plot(x="acc_time", y="epoch", ax=ax, label="RN34", c="blue", linewidth=5, alpha=0.3)

#df_sin_18_40p.plot(x="acc_time", y="epoch", ax=ax, label="RN18, 40%", c="red", linestyle='dashed', linewidth=2, alpha=0.5)
#df_sin_34_60p.plot(x="acc_time", y="epoch", ax=ax, label="RN34, 60%", c="blue", linestyle='dashed', linewidth=2, alpha=0.5)

#df_mul_18_50p.plot(x="acc_time", y="epoch", ax=ax, label="RN18, shared data", c="red", linestyle='dashed', linewidth=5, alpha=0.5)
#df_mul_34_50p.plot(x="acc_time", y="epoch", ax=ax, label="RN34, shared data", c="blue", linestyle='dashed', linewidth=4, alpha=0.5)

df_mul_18_40p.plot(x="acc_time", y="epoch", ax=ax, label="RN18, shared data, 40%", c="red", linestyle='dotted', linewidth=2, alpha=0.3)
df_mul_34_60p.plot(x="acc_time", y="epoch", ax=ax, label="RN34, shared data, 60%", c="blue", linestyle='dotted', linewidth=3, alpha=0.3)

df_mul_18_40p_q20.plot(x="acc_time", y="epoch", ax=ax, label="RN18, shared data, 40%, q=20", c="red", linestyle='dashdot', linewidth=5, alpha=0.8)
df_mul_34_60p_q20.plot(x="acc_time", y="epoch", ax=ax, label="RN34, shared data, 60%, q=20", c="blue", linestyle='dashdot', linewidth=4.5, alpha=0.8)

plt.yticks([i for i in range(1, 11)])
plt.xlim((0,900))
plt.ylim((1,10))
plt.legend(loc="upper left")
plt.ylabel("Epochs")    
plt.xlabel("Elapsed seconds")
plt.suptitle("One dataloader per training instance vs. \nUsing one dataloader with queue (with/without MPS)", y=0.93, fontsize=16)
#plt.title("Epoch numbers annotated", fontsize=10)
plt.grid()
plt.show()
```
